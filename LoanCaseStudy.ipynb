```python
l_lam_data = ldata.dropna(subset=['LoanAmount'])
l_lam_data
```

```python
l_lam_data.isna().sum()
```

```python
l_lam_data['Gender'].fillna(l_lam_data['Gender'].mode()[0], inplace=True)
l_lam_data['Married'].fillna(l_lam_data['Married'].mode()[0], inplace=True)
```

```python
l_lam_dum = pd.get_dummies(data=l_lam_data, columns=['Gender','Married','Property_Area','Education'],drop_first=True)
```

```python
l_lam_dum.isna().sum()
```

```python
l_lam_dum.dtypes
```

```python
l_lam_dum.Self_Employed.replace(('Yes', 'No'), (1, 0), inplace=True)
```

```python
l_lam_dum.isna().sum()
```

```python
plt.subplots(figsize=(20,15))

sns.heatmap(l_lam_dum.corr(),annot=True)
```

```python
l_lam_dum.columns
```

```python
l_selfem = l_lam_dum[['ApplicantIncome','LoanAmount','Property_Area_Urban','Self_Employed']]
```

```python
l_selfem = l_selfem.dropna(subset=['Self_Employed'])
```

```python
l_selfem_x = l_selfem[['ApplicantIncome','LoanAmount','Property_Area_Urban']]
```

```python
l_selfem_y=l_selfem[['Self_Employed']]
```

```python
from sklearn.preprocessing import StandardScaler
stdScaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(l_selfem_x,l_selfem_y)
```

```python
stdScaler.fit(x_train)
```

```python
x_train_std = stdScaler.transform(x_train)
```

```python
x_test_std = stdScaler.transform(x_test)
```

```python
x_train_std
```

```python
from sklearn.linear_model import LinearRegression
seModel = LinearRegression()
```

```python
seModel.fit(x_train_std, y_train)
```

```python
seModel.score(x_train_std,y_train)
```

```python
seModel.score(x_test_std,y_test)
```
