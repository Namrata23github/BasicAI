```python
import statsmodels.api  as sm
```

```python
x=hdata.SqFt
y=hdata.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
regmodel = sm.OLS(y_train, x_train)
```

```python
results = regmodel.fit()
```

```python
print(results.summary())
```

```python
y_train_pred=results.fittedvalues
```

```python
residual = y_train-y_train_pred
```

```python
ax = sns.scatterplot(x=y_train_pred, y=residual)
```

```python
import scipy as sp
```

```python
fig,ax = plt.subplots(figsize=(6,2.5))
sp.stats.probplot(residual,plot=ax,fit=True)
```

```python
y_test_pred =results.predict(x_test)
residul_test = y_test-y_test_pred
```

```python
fig,ax = plt.subplots(121)
ax = sns.scatterplot(x=y_test_pred, y=residul_test)
sm.qqplot(residul_test)
```
